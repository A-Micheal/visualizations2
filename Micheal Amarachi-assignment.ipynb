```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
tipsdata=sns.load_dataset('tips')
tipsdata
```

```python
#Checking for missing values
tipsdata.isnull().sum()
```

```python
tipsdata.describe()
```

```python
#Plotting joint plot of total bill against tip by sex
sns.jointplot(
    data=tipsdata,
        x="total_bill",
        y="tip",
        hue ="sex",
        kind="scatter",
        palette="Set2"
    )
plt.suptitle(f"Total Bill vs Tip by Sex", fontsize=14)
plt.tight_layout()
plt.show()
    
```

## **OBSERVATIONS FROM THE CHART**
1. There seems to be a positive correlation between the tips and total bill
as the total bill increases the tip increases.    
2. Male customers tend to tip more then female customers.
3. Most bills are clustered between 10 and 25 dollars.
